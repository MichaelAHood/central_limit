```python
# Now, let's see if I can figure out a function 
# that will simulate a coin toss for n times and 
# return the proportion of total tosses that are heads

def experiment(n):
    
    result = {"heads": 0.0, "tails" : 0.0}
    
    for i in range(n):
        
        cointoss = random.randrange(0,2)
        
        if cointoss == 0: # simulate a heads condition
            result['heads'] += 1
        else:
            result['tails'] += 1 # simultae a tails condition
       
    #print result['heads'], result['tails']
    
    proportion = result['heads'] / n
    
    return proportion
```

```python
# let's put the outcomes in some kind of data structure for plotting

outcomes = []

for i in range(0, 10): 
    outcomes.append(experiment(100))
```

```python
# we need to use a series data struct for plot

outcomes_series = pd.Series(outcomes)
```

```python
figure = outcomes_series.plot(kind='hist', alpha=0.5, bins=20)
```

```python
# Now define a procedure that runs many experiments and 
# also returns the outcomes of those many experiemnts
# as a series data struct.

def runManyExperiments(num_experiments, size_experiments):
    
    outcomes = []
    
    for i in range(num_experiments):
        outcomes.append(experiment(size_experiments))
    
    return pd.Series(outcomes)
```

```python
# Now let's see if it works

outcomes = runManyExperiments(10, 100)

figure = outcomes.plot(kind='hist', alpha=0.5, bins=20)
```

```python
# Now, let's try a hundred experiments and see what pops out

outcomes = runManyExperiments(100, 100)

figure = outcomes.plot(kind='hist', alpha=0.5, bins=20)
```

```python
# Ok, that looks better.

# Let's increase the number of coin tosses per experiment to 1,000 and see if 
# we can make this plot start to look something like a bell curve.  

# I should also increase the number of bins to dispaly a wider variation
# in experiemnt results.

outcomes = runManyExperiments(100, 1000)

figure = outcomes.plot(kind='hist', alpha=0.5, bins=50)
```

```python
# I like where this is going
# Let's try making the experiemnt size and number 10 times larger

outcomes = runManyExperiments(1000, 10000)

figure = outcomes.plot(kind='hist', alpha=0.5, bins=50)
```
